# Correlation heatmap of Student Performance

## Import the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the data

```python
df = pd.read_csv('/content/StudentPerformance.csv')
```

## Encode Categorical Data

```python
# Encode categorical column: Yes → 1, No → 0
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes': 1, 'No': 0})

# Quick check
print(df[['Extracurricular Activities']].head())
```

## Create the Correlation Matrix

```python
# Assuming df is your cleaned dataset with numeric columns
corr_matrix = df.corr()

# Display the correlation matrix
print(corr_matrix.round(2))
```

## Heatmap Visualization

```python
# Assuming corr_matrix is already created
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,
            annot=True,        # show correlation values
            cmap='coolwarm',   # color scheme: warm = strong, cool = weak
            center=0)          # center at 0 for balance
plt.title("Correlation Heatmap of Student Performance Data")
plt.show()
```

## Notes
How to Read the Heatmap
1. Focus on One Column
Start with the Performance Index column (your target outcome).

Ignore the rest for now — you only care about how each variable relates to performance.

2. Interpret the Values
High positive correlation (close to +1) → strong predictor of performance.

Moderate correlation (around 0.3–0.6) → useful, but not the strongest.

Near 0 → almost no linear relationship.

Negative correlation (close to -1) → inverse relationship (rare in your dataset, but possible).

3. Typical Findings in Student Data
Previous Scores → Strong positive Students who did well before tend to perform well again.

Hours Studied → Moderate positive More study hours generally improve performance, but not as strongly as past scores.

Sample Papers Practiced → Moderate positive Practice helps, but effect size is smaller than raw study hours or prior scores.

Sleep Hours → Weak positive Sleep matters, but correlation is usually low because it’s influenced by many other factors.

Extracurricular Activities → Very weak or neutral Participation doesn’t show a strong direct link to performance (though it may help indirectly with skills like discipline).

4. Why This Reading Strategy Works
By isolating Performance Index, you avoid being overwhelmed by the full matrix.

You immediately see which variables are worth attention and which can be ignored in predictive modeling.

This is the bridge between EDA (exploratory data analysis) and feature selection for machine learning.
